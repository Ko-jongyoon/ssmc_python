```python
import pandas as pd
import numpy as np
```

```python
data={'kor':[80,30,50,70],'eng':[20,90,90,80],
     'math':[33,44,55,66]}  #키값이 리스트인 경우
df =pd.DataFrame(data,index=['aa','bb','cc','dd'])
df
```

```python
# col 추가
df['sci'] =[1,2,3,4] # col값이 있으면 값변경, 없으면 col추가
df
```

```python
df['sci']= df['kor']+2 # 'sci' col에 'kor' col 값 +2의 값을 준다
df
```

```python
df['kor']+df['eng']
```

```python
df['sum'] =df['kor']+df['eng']+df['math']
df
```

```python
df
```

```python
# row 추가
df.loc['ee']=[1,2,3] # row값이 있으면 수정 , 없으면 추가 # row 값 추가
#df.loc['dd']=[1,2,3] # row 값 수정
df
```

```python

df.loc['bb':,'aaa'] =[2,3,4]  #loc,iloc을 통해 특정범위 추가 
df.loc['cc':,'bbb'] =[3,4]
df
```

```python
df1=df
# df1.dropna() # nan을 삭제하는 메소드 # nan이 있는 row를 삭제하게 된다
# df1.dropna(axis=1) #nan 이있는 col을 삭제하게된다
# df1.dropna(subset=['aaa']) # 'aaa'에대해서 nan이 들어간 row를 전부 삭제
#df1.dropna(subset=['aaa', 'bbb'])
#df1.fillna(0) # nan데이터를 모두 0으로 다 채워줌
df1['aaa'].fillna(0, inplace=True) # 'aaa'에대해서 nan의 값을 0으로 채움
df1
```
