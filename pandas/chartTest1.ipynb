```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt     
import matplotlib

matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
# matplotlib 함수를 살펴보자
# y축만 주게되면 x는 기본적으로 0,1,2,3,4로 정의된다 data1은 y축이다
data1 = [10,14,19,20,25]
plt.plot(data1)
plt.show()
```

```python
# y축만 주게되면 x는 기본적으로 0,1,2,3,4로 정의된다
y = [10,14,19,20,25]
x = [1,2,3,4,5]
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
# plt.figure() # 하나의 윈도우를 만들고 그윈도우에 라인을 그려라  이것은 생략되어있다 결과는 똑같다 
plt.plot(x,y1)
plt.figure(figsize=[3,3]) # 두개의 윈도우에 라인 차트가 그려진다 피규어를 선언해주지않으면 , 하나를 만들어서 거기에 다 때려넣는다.
plt.plot(x,y2)            # figsize = 인치 단위
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
# plt.figure() # 생략되어있다. 
plt.subplot(2,2,1) #  하나의 윈도우를 2행 2열로 분할했을때 첫번째
plt.plot(x,y1)
plt.subplot(2,2,2) # 2행 2열로 분할했을때 두번째
plt.plot(x,y2)
plt.subplot(2,2,3) # # 2행 2열로 분할했을때 세번째
plt.plot(x,y3)
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
# plt.figure() # 생략되어있다. 
plt.subplot(3,1,1)  # 3행 1열에서 첫번째 
plt.plot(x,y1)
plt.subplot(3,1,2) 
plt.plot(x,y2)
plt.subplot(3,1,3) 
plt.plot(x,y3)
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]   
plt.ylim(0,30)         # default는 y1의 가장 작은값부터 가장 큰값으로 정의 ,# ylim은 강제로 y축 범위 설정 가능
plt.xlim(0,10)         
plt.plot(x,y1) 
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]   
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
plt.xlabel("x축라벨")
plt.ylabel("y축라벨")
plt.title('타이틀')
plt.grid(True)
plt.plot(x,y1, 'ro:', x,y2,'b>-', x,y3,'g1-') # r= color, o = marker , : = Line '123'
plt.legend(['국어','영어','수학'])
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]   
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
plt.xlabel("x축라벨")
plt.ylabel("y축라벨")
plt.title('타이틀')
plt.grid(True)
plt.plot(x,y1, 'ro--') 
plt.legend(['국어'])
for x,y in zip( x,y1 ):
    plt.text(x,y,'%d점'%y)
# plt.text(1,10,'10점')
# plt.text(2,14,'14점')
plt.show()
```
