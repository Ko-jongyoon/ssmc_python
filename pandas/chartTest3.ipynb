```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt     
import matplotlib

matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
x = ['홍길동','이순신','임꺽정','이이']
# x= x.sort()  # 문자열은 sort() 사용해야 순서대로 졍렬가능
print( sorted(x) ) # x축이 문자열이면 정렬이 안되고 그대로 나오기때문에 x축이 list인경우 sorted 
y = [90,77,50,80]
plt.bar( sorted(x),y)
plt.show()
```

```python
x = pd.Series(['홍길동','이순신','임꺽정','이이'])  # x 축이 series인 경우
y = [90,77,50,80]
plt.bar( x.sort_values(), y)
plt.show()
```

```python
x = ['홍길동','이순신','임꺽정','이이']
y = [90,77,50,80]
plt.bar( x,y, color=['r','g','b','c'])
plt.show()
```

```python
x = ['홍길동','이순신','임꺽정','이이']
y = [90,77,50,80]
plt.barh( x,y, color=['r','g','b','c'])
```

```python
xLabel = ['홍길동','이순신','임꺽정','이이']
x = pd.Series([1,2,3,4])  
kor =[90,77,50,80]
eng =[60,100,50,20]
y = [90,77,50,80]
plt.xticks(x+0.2,xLabel) # x축을 xLabel로 바꿈
plt.bar( x,kor, color='r', width=0.4, label="국어")
plt.bar( x+0.4,eng, color='c', width=0.4,label="영어") # +0.5 위치조정 (series여야만 가능하다) 즉 x+0.5는가능 하나 xLabel+0.5는 불가
plt.xlabel('이름')
plt.ylabel('점수')
plt.title('1학년성적')
plt.legend() # plt.legend(['국어','영어'])
plt.show()
```

```python
colors = ['r','g','b','y']
xLabel = ['홍길동','이순신','임꺽정','이이']
x =[1,2,3,4]
kor = [90,80,87,70]
eng = [80,70,80,90]

plt.bar(x,kor,color='r',
        width=0.4,
        label='국어',
        alpha = 0.3)

plt.bar(x,eng,color='b',
        width=0.4,
        label='영어', 
        bottom=kor,# 영어 점수를 국어 점수 밑에 깔고 더해라 -> 총점도 잘 보일듯
        alpha = 0.3
       )

for x_,k,e in zip(x,kor,eng):
    plt.text(x_,k/2,'%.2f\n(%d점)'%((k/(k+e)),k),va = 'center', ha = 'center')
    plt.text(x_,e/2+k,'%.2f\n(%d점)'%((e/(k+e)),e),va = 'center', ha = 'center')
    
plt.xticks(x,xLabel)
plt.xlabel('이름')
plt.ylabel('점수')
plt.title('1학년 성적')
plt.legend()
plt.show()
```
