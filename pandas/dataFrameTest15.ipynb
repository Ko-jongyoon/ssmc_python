```python
from bs4 import BeautifulSoup
import urllib.request as REQ
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt     
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
urlbase= 'http://www.menupan.com/'
urlsub = '/restaurant/bestrest/bestrest.asp?pt=rt&areacode=ss201'

response= REQ.urlopen(urlbase + urlsub)
soup = BeautifulSoup(response,'html.parser')
```

```python
menuList=[]
hrefList = []
ul = soup.select_one('.rankingList .list') # 리스트 0번째만
for li in ul.select('li'):
    href = li.select_one('.listName a')['href']
    hrefList.append(href)

    rank = li.select_one('.numTop, .rankNum').string
    listName = li.select_one('.listName').string
    listType = li.select_one('.listType').string
    listArea = li.select_one('.listArea').string
    menuList.append({"랭크":rank,
                     "상호명":listName,
                     "업종":listType,
                     "지역":listArea})
```

```python
hrefList
```

```python
df = pd.DataFrame(menuList)
```

```python
df
```

```python
telList=[]
addList=[]
urlbase= 'http://www.menupan.com'
for href in hrefList:
    response= REQ.urlopen(urlbase + href)
    soup = BeautifulSoup(response,'html.parser')
    telList.append(soup.select_one('.tel1').string)
    addList.append(soup.select_one('.add1').string)
    print(soup.select_one('.tel1').string)
    print(soup.select_one('.add1').string)
    
```

```python
df['전화번호'] =telList
df['주소'] =addList
df
```
