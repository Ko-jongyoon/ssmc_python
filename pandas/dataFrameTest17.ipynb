```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt     
import matplotlib

matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
employee = [{'empno':1 ,'ename':'kim', 'dept':1},
            {'empno':2 ,'ename':'lee', 'dept':2},
            {'empno':3 ,'ename':'park', 'dept':1},
            {'empno':4 ,'ename':'song', 'dept':3},
            {'empno':5 ,'ename':'min', 'dept':2}]

dept =[{'dept':1,'deptname':'관리직'},
       {'dept':2,'deptname':'영업직'},
       {'dept':3,'deptname':'개발직'},]

info =[{'empno':1,'addr':'서울시','phone':'010-1111-1111'},
       {'empno':2,'addr':'부산시','phone':'010-2222-2222'},
       {'empno':3,'addr':'광주시','phone':'010-3333-3333'},
       {'empno':4,'addr':'광주시','phone':'010-4444-4444'},
       {'empno':5,'addr':'광주시','phone':'010-5555-5555'}]
```

```python
dfemp=pd.DataFrame(employee)
dept =pd.DataFrame(dept)
info =pd.DataFrame(info)
```

```python
dfemp
```

```python
m = pd.merge(dfemp,dept, on='dept') # merge는 결합시켜주는 판다스 함수다.
m.drop('dept',axis=1,inplace=True)
m
```

```python
pd.merge( m, info, on='empno')
```

```python
#pd.concat( [dfemp , dfemp] ,axis=1)  # axis 을이용하여 row와 col 결정
pd.concat( [dfemp , dfemp] ,axis=0)   
```

```python
# pd.concat( [dfemp , dept] ,axis=1, join ='outer') # 없는 부분은 NaN처리  
pd.concat( [dfemp , dept] ,axis=1, join ='inner')  #   inner = 5개짜리와 4개짜리가 붙으면 작은 4개짜리를 보여준다
                                                   #   outer = 5개짜리와 4개짜리가 붙으면  큰 5개 짜리를 보여준다.
```
