```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt     
import matplotlib

matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
cctv =pd.read_csv('CCTV_in_Seoul.csv')
cctv.head()
```

```python
pop = pd.read_excel('population_in_Seoul.xls',header=2, usecols='B,D,G,J,N') # header =2 : 2번째 col을 헤더로 주겠다. 
                                                                         # usecols=  :필요한 col만을 선택해서 보여준다.
pop.columns = ['기관명','인구총합','내국인총합','외국인총합','65세 이상']  # col명 변경
pop.head()
```

```python
pop = pop.dropna() # NAN 데이터 제거 함수 
pop
```

```python
cctv['증가율']= (cctv['2014년'] + cctv['2015년']+ cctv['2016년'])/cctv['2013년도 이전']
cctv =cctv[['기관명','소계','증가율']]
cctv
```

```python
cctv.rename( columns={'소계':'cctv총수'},inplace=True)
cctv
```

```python
df = pd.merge(cctv,pop, on='기관명')
df.head()
```

```python
df.set_index('기관명',inplace=True)
```

```python
df.head()
```

1. cctv가 가장많은 순으로 정렬하시오

```python
df.sort_values('cctv총수', ascending=False)['cctv총수']
```

2. 기관별 cctv현황을 바차트(수평)으로 표시하시오

```python
x = df.index
plt.barh(x,df['cctv총수'],color='r')
plt.ylabel('기관명')
plt.xlabel('cctv총수')
plt.legend(['CCTV수'])
```

3.기관별 cctv비율을 바차트(수평)로 표시하시오

```python
x = df.index
plt.barh(x,df['증가율'],color='r')
plt.ylabel('기관명')
plt.xlabel('증가율')
plt.legend(['CCTV수'])
```

4. 구별 외국인 비율과 고령자비율 컬럼을 추가하시오

```python
df['외국인비율']= df['외국인총합']/df['인구총합']
df['고령자비율']= df['65세 이상']/df['인구총합']
df
```

#5 기관별 인구 대비 cctv총수에 cctv비율 column을 추가하시오

```python
df['cctv 비율'] = df['cctv총수']/df['인구총합']
df
```

    #6  기관별 cctv비율을 bar chart(수평)으로 표시하시오.

```python
df1 =df['cctv 비율']
plt.barh( df.index, df1 , alpha=0.3)
plt.show()
```
